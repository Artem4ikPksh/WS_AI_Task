```python
# EDA for animal dataset
import os
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import random
import tensorflow as tf

# Path to the dataset
DATA_DIR = "animal_data"

# Checking classes
classes = os.listdir(DATA_DIR)
print("Number of classes:", len(classes))
print("List of classes:", classes)

# Counting the number of images in each class
class_counts = {}
for cls in classes:
    cls_path = os.path.join(DATA_DIR, cls)
    images = os.listdir(cls_path)
    class_counts[cls] = len(images)

# Output statistics
print("Number of images by class:", class_counts)

#  Visualization of class distribution
plt.figure(figsize=(10,6))
sns.barplot(x=list(class_counts.keys()), y=list(class_counts.values()))
plt.xticks(rotation=45)
plt.title("Distribution of images by class")
plt.show()

#  Visualization of image examples
plt.figure(figsize=(12,8))
for i, cls in enumerate(random.sample(classes, 5)):  # випадкові 5 класів
    cls_path = os.path.join(DATA_DIR, cls)
    img_name = random.choice(os.listdir(cls_path))
    img_path = os.path.join(cls_path, img_name)
    
    img = tf.keras.utils.load_img(img_path, target_size=(128,128))
    plt.subplot(1,5,i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(cls)
plt.show()

# Check image sizes (whether there are different formats)
sizes = []
for cls in classes:
    cls_path = os.path.join(DATA_DIR, cls)
    for img_name in os.listdir(cls_path)[:10]:  # check the first 10 images in each class
        img_path = os.path.join(cls_path, img_name)
        img = tf.keras.utils.load_img(img_path)
        sizes.append(img.size)

print("Examples of image sizes:", sizes[:10])
```
